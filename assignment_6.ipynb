```python
#17
d=pd.Series((df.loc[:,'RainToday']))
y=0
n=0

for x in range (0,366):
    if d[x]=='Yes':
        y=y+1
    else:
        n=n+1
rain_label=['Yes','No']
rain_data=[y,n]
plt.pie(rain_data,labels=rain_label)
plt.title('Rain in a year')
plt.show()

```

```python
#18
plt.bar(rain_label, rain_data)
plt.ylabel('No. of days')
plt.xticks(fontsize=15, rotation=30)
plt.title('Rain In a year')
plt.show
```
